### 문자열 인덱싱 슬라이싱
- 문자열도 인덱스 번호를 가진다
- 인덱싱 : 인덱스 번호 하나를 가리켜서 데이터 추출
- 슬라이싱 : 범위를 지정하여 여러 개의 데이터 추출

```python
name = "My name is SJ"
```

```python
print(name[3])
```

```python
name[-1::-1]
```

```python
# 슬라이싱 : 변수명[ 범위 ]
# 변수명[시작인덱스 : 끝인덱스 + 1]
# 주의1: 슬라이싱에서 범위를 작성할 때 끝인덱스 범위는 내가 가져오고싶은 인덱스 + 1

# My
# name
# YH
print(name[:2])
print(name[3:7])
print(name[-2:])
```

### 문자열 포매팅

```python
month = 6
day = 25
today = "오늘은 " + month + "월" + day + " 일입니다."
today
```

```python
# 1. % 기호포매팅 -> % 기호를 이용하여 숫자 대입
# 단점 : 데이터타입에 따른 알파벳을 암기
month = 6
day = 25
today = "오늘은 %d월 %d일 입니다"%(month, day)
today
```

```python
print(f' 오늘은 {month}월 {day}일입니다.')
```

```python
# 2. format 함수
#  중괄호 순서와 포맷함수 안에 들어가는 변수 순서가 일치
month = 6
day = 25
today = "오늘은 {}월 {}일 입니다".format(month, day)
today
```

```python
# 3. f-string
month = 6
day = 25
today = f'오늘은 {month}월 {day}일입니다'
today
```

### 연산자

```python
num1 = int(input("정수1 입력 >> "))
num2 = int(input("정수2 입력 >> "))
print(f'더하기 결과 : {num1 + num2}')
print(f'빼기 결과 : {num1 - num2}')
print(f'곱하기 결과 : {num1 * num2}')
print(f'나누기 결과 : {num1 / num2}')
```
