```python
num1, num2 = map(int, input().split())
if num1 > num2:
    print("첫 번째 정수가 더 큽니다.")
elif num1 < num2:
    print("두 번째 정수가 더 큽니다.")
else:
    print("두 수가 똑같습니다.")
```

```python
sc = int(input("점수 입력"))
grade = ''
if sc >= 90:
    grade = 'A'
elif sc >= 80:
    grade = 'B'
elif sc >= 70:
    grade = 'C'
elif sc >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f'{sc}점은 {grade}학점입니다.')
```
