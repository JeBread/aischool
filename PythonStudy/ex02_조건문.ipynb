```python
sc = int(input(" "))
grade = ''
if sc >= 90:
    grade = 'A'
elif sc >= 80:
    grade = 'B'
elif sc >= 70:
    grade = 'C'
elif sc >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f'{sc} {grade}.')
```
