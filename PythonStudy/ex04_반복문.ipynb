```python
import random as rd
rdNum = rd.randint(1, 50)
while True:
    print("  >> ")
    num = int(input())
    if num == rdNum:
        print("")
        break
    elif num < rdNum:
        print(f'{num}   .')
    else:
        print(f'{num}   .')
    
```
