```python
import random as rd
rdNum = rd.randint(1, 50)
while True:
    print("숫자를 입력해주세요 >> ")
    num = int(input())
    if num == rdNum:
        print("정답입니다")
        break
    elif num < rdNum:
        print(f'{num}보다 큰 수 입니다.')
    else:
        print(f'{num}보다 작은 수 입니다.')
    
```

```python
scores = [90, 45, 70, 60, 55]
for i in range(len(scores)):
    if scores[i] >= 60:
        print(f'{i+1}번 학생은 합격입니다.')
    else:
        print(f'{i+1}번 학생은 불합격입니다.')
```
