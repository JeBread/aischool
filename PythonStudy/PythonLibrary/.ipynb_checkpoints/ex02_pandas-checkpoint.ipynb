```python
# 라이브러리 불러오기~
import pandas as pd
```

```python
# Series -> pandas 자료구조 중 하나로 인덱스와 값이 1:1로 대응하는 1차원 자료구조
population = pd.Series([9668465,3391946,2942828,1450062],
                      index = ['서울','부산','인천','광주'])
population
```

```python
# Series 값만 확인하기
population.values
```

```python
# Series 인덱스만 확인
population.index
```

```python
# Series 이름 지정하기
population.name = "인구"
population
```

### 인덱싱 슬라이싱

```python
# 부산의 인구수 출력
population["부산"]
```

```python
# 슬라이싱 부산 ~ 광주
population["부산":"광주"]
```
