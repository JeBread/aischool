```python
#      
def num_sum(num1, num2):
    return num1 + num2
num1, num2 = map(int, input().split())
print(num_sum(num1, num2))
```

```python
a, b = map(int, input("      . ").split())
op = input(" (+ or -)")
def cal(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    else:
        return "  "
res = cal(a, b, op)
print(f' : {res}')
    
```

```python
def all_sum(*args): #arguments
    sum = 0
    for x in args:
        sum += x
    return sum
print(all_sum(1, 4, 5, 7))
```
