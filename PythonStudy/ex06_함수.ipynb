```python
# 두 개의 정수를 더해주는 함수 정의
def num_sum(num1, num2):
    return num1 + num2
num1, num2 = map(int, input().split())
print(num_sum(num1, num2))
```

```python
a, b = map(int, input("빈 칸으로 구분해서 두 개의 정수를 입력해주세요. ").split())
op = input("연산자 입력(+ or -)")
def cal(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    else:
        return "연산자 똑바로 쳐"
res = cal(a, b, op)
print(f'결과 : {res}')
    
```

```python
def all_sum(*args): #arguments
    sum = 0
    for x in args:
        sum += x
    return sum
print(all_sum(1, 4, 5, 7))
```

### 함수의 기본값

```python
# 함수를 정의할 때 기본값을 지정할 수 있음
# 숫자데이터에 제곱을 계산해주는 함수
def power_num(num1, power = 2): # power를 입력해주지 않으면 기본 2
    return num1 ** power
power_num(4)
```

```python
# 속성 변경 -> 속성명 = 값
power_num(3, power = 3)
```
